````python
synonym_names_df = pd.read_excel('../mood_tags.xlsx', index_col=0, sheet_name='Mood Emotion')
````

````python
synonym_dict = {}
for index, data in synonym_names_df.iterrows():
    if not pd.isnull(data['Synonym(s)']): # There are rows with no data. We don't need to process them.
        value = index.lower().strip() # lowercase word, and removed white spaces
        
        # Split the word list by comma(,), then for each word => lowercase word, and removed white spaces
        keys = [ s.lower().strip() for s in data['Synonym(s)'].split(',')] 
        
        d = {key: value for key in keys } # Assign value to  list of keys
        
        synonym_dict.update(d)
    
````

````python
synonym_dict
````

````python
remove_tags_df = pd.read_excel('../mood_tags.xlsx', sheet_name='Remove ') # Read the remove tag dataframe
case_remove_tags = remove_tags_df['Remove'].tolist() # Convert the tags into python list
remove_tags = [tag.lower().strip() for tag in case_remove_tags] # lower case and remove white spaces
````

````python
remove_tags
````

### This code prunes for the remove tags, null values and replace with synonyms

````python

Y = []
for row in raw_Y: # Iterate over each row
    y = []
    for element in row: # Iterate over each element in the row
        if not pd.isnull(element) and element.lower().strip() not in remove_tags: #Filter based on condtion
            element = element.lower().strip()
            syn = synonym_dict.get(element, element) # get the synonym word from dict
            y.append(syn)
    y = list(set(y))
    Y.append(y)
````

````python
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
X_train, X_test, y_train, y_test = train_test_split(X, multi_label_Y, test_size=0.01, random_state=42)
````

````python
#model = ridge_classifier_cv.fit(X, multi_label_Y)
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score

# For this kind of classification problem: We need to use F1 score for the model accuracy.
def get_initial_result(model, normalization=None):
    _X = copy(X)
    if normalization is not None:
        _X = normalization.transform(_X)
    result = model.predict(_X)
    cnt = 0
    for (index, _), pred_res in zip(df.iterrows(), result):
        print(index, '   >>>>>>  ', mlb_y.inverse_transform(np.array([pred_res])))
        if cnt > 10:
            break
        cnt += 1
        
def get_scores(model, normalization=None):
    _X = copy(X_test)
    if normalization is not None:
        _X = normalization.transform(_X)
    preds = model.predict(_X)
    # score = classification_report(preds, y_test)
    score = f1_score(preds, y_test, average='macro')
    print(score)
    
def train_model(x_train, y_train, classifier=linear_model.RidgeClassifierCV(cv=10)):
    model = OneVsRestClassifier(classifier).fit(x_train, y_train)
    return model

def train_and_report(x_train,y_train, classifier):
    model = train_model(x_train,y_train, classifier)
    get_scores(model)
    get_initial_result(model)
    return model
    
    
````

### Taking Feature Imortance to Caculate the Euclidean Distance

````python
avg_feature_weight = ridge_classifier_cv_model_OneVsRestClassifier.coef_.mean(axis=0)
# normalized_avg_feature_weight  = preprocessing.normalize(avg_feature_weight) # Normalize the feature importnace asbove?
normalized_avg_feature_weight = avg_feature_weight / np.sqrt((np.sum(avg_feature_weight**2)))
````

````python
normalized_avg_feature_weight
````

## For the Euclidean Distance We have used the RidgeClassifer

````python
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.neighbors import DistanceMetric

dist = DistanceMetric.get_metric('euclidean')


def euclidean(vector1, vector2):
    ''' use numpy.linalg.norm to calculate the euclidean distance. '''
    distance = np.linalg.norm(vector1-vector2, 2, 0) # the third argument "0" means the column, and "1" means the line.
    return distance

def get_pairwise_distance(song1, song2):
    feature1 = df[feature_columns].loc[song1].as_matrix()
    feature2 = df[feature_columns].loc[song2].as_matrix()
    
    w1_feature1  = normalized_avg_feature_weight * feature1
    w2_feature1  = normalized_avg_feature_weight * feature2
    
    m = np.array([w1_feature1, w2_feature1])
    val = dist.pairwise(m)
    #     norm_val = preprocessing.Normalizer().fit_transform(dist.pairwise(m))
    
    return val

def get_top_n_similar_songs(iloc):
    feature1 = scaler_norm_x[iloc]
    score_dict = {}
    for index, feature2 in enumerate(scaler_norm_x):
        w1_feature1  = normalized_avg_feature_weight * feature1
        w2_feature1  = normalized_avg_feature_weight * feature2
        m = np.array([w1_feature1, w2_feature1])
        #distance = dist.pairwise(m)[0][1]
        distance = euclidean(w1_feature1, w2_feature1)
        score_dict[list_of_songs[index]] = distance
    sorted_list = sorted(score_dict.items(), key=lambda x: x[1])
    return sorted_list
        

    
````

```
From the list of 3984 tracks in the dataset shuffle and extract 150 tracks at random.
These will be the reference tracks.

For each 150 reference tracks extract:
1st closest track
2nd closest track
3rd closest track
1200 closest track
2300 closest track
3984 closest track (actually the furthest away)

This will make for 150 x 7 = 1050 tracks in total
As a result I am looking for a ranking list with the track names for 150 pools 
Pool1 : Reference track, 1st closest, 2nd closest etc...
Pool2: Reference track, 1st closest, 2nd closest etc...
Pool3: Reference track, 1st closest, 2nd closest etc...
And so on...
```

````python
import random
random_150_numbers = [random.randint(0, 3984) for i in range(150) ]
result_df = []
def get_sim_songs():
    result_row = []
    random_150_numbers = [random.randint(0, 3984) for i in range(150) ]
    for rand_song_index in tqdm.tqdm(random_150_numbers, total=len(random_150_numbers)):
        sorted_score_list = get_top_n_similar_songs(rand_song_index)
        song = list_of_songs[rand_song_index]
#         result_row = [song, sorted_score_list[1][0], sorted_score_list[2][0], sorted_score_list[3][0], sorted_score_list[1200][0], sorted_score_list[2300][0], sorted_score_list[-1][0]]
        result_row = [song, sorted_score_list[1][0], sorted_score_list[2][0], sorted_score_list[3][0], sorted_score_list[1200][0], sorted_score_list[2300][0], sorted_score_list[-1][0]]
        result_df.append(result_row)
    return pd.DataFrame(result_df, columns='Reference Song,1st closest track,2nd closest track,3rd closest track,1200th closest track,2300th closest track,3983rd closest track'.split(','))       
````
