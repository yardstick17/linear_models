```python
synonym_names_df = pd.read_excel('../mood_tags.xlsx', index_col=0, sheet_name='Mood Emotion')
```

```python
synonym_dict = {}
for index, data in synonym_names_df.iterrows():
    if not pd.isnull(data['Synonym(s)']): # There are rows with no data. We don't need to process them.
        value = index.lower().strip() # lowercase word, and removed white spaces
        
        # Split the word list by comma(,), then for each word => lowercase word, and removed white spaces
        keys = [ s.lower().strip() for s in data['Synonym(s)'].split(',')] 
        
        d = {key: value for key in keys } # Assign value to  list of keys
        
        synonym_dict.update(d)
    
```

```python
synonym_dict
```

```python
remove_tags_df = pd.read_excel('../mood_tags.xlsx', sheet_name='Remove ') # Read the remove tag dataframe
case_remove_tags = remove_tags_df['Remove'].tolist() # Convert the tags into python list
remove_tags = [tag.lower().strip() for tag in case_remove_tags] # lower case and remove white spaces
```

### This code prunes for the remove tags, null values and replace with synonyms

```python

Y = []
for row in raw_Y: # Iterate over each row
    y = []
    for element in row: # Iterate over each element in the row
        if not pd.isnull(element) and element.lower() not in remove_tags: #Filter based on condtion
            element = element.lower().strip()
            syn = synonym_dict.get(element, element) # get the synonym word from dict
            y.append(syn)
    y = list(set(y))
    Y.append(y)
            
```

### Taking Feature Imortance to Caculate the Euclidean Distance

```python
avg_feature_weight = scaler_norm_model.coef_.sum(axis=0)
```

```python
avg_feature_weight.shape
```

```python
normalized_avg_feature_weight  = preprocessing.scale(avg_feature_weight)
```

```python
normalized_avg_feature_weight
```

```python
df.head()
```

```python
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.neighbors import DistanceMetric
dist = DistanceMetric.get_metric('euclidean')
def get_pairwise_distance(song1, song2):
    feature1 = df[feature_columns].loc[song1].as_matrix()
    feature2 = df[feature_columns].loc[song2].as_matrix()
    w1_feature1  = normalized_avg_feature_weight * feature1
    w2_feature1  = normalized_avg_feature_weight * feature2
    m = scaler.transform(np.array([w1_feature1, w2_feature1]))
    return dist.pairwise(m)
```
