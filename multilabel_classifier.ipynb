# Filter the label columns based on column names. The 'mood metadata' is a prefix for lable column.

```python
label_columns = [col for col in df.columns if 'mood metadata' in col]
label_columns
```

## Get the features from the dataframe using the featue colums

```python
features = df[feature_columns] 
```

## Get the Labels from the dataframe using the label colums


```python
labels = df[label_columns]
labels
```

## Get the feature matrix from the dataframe

```python
X = features.as_matrix()
```

```python
raw_Y = [list(row) for row in labels.as_matrix()]
```

```python
raw_Y
```

```python
sheets = pd.ExcelFile('../mood_tags.xlsx')
sheets.sheet_names
```

```python
synonym_names_df = pd.read_excel('../mood_tags.xlsx', index_col=0, sheet_name='Mood Emotion')
```

```python
synonym_dict = {}
for index, data in synonym_names_df.iterrows():
    if not pd.isnull(data['Synonym(s)']): # There are rows with no data. We don't need to process them.
        value = index.lower().strip() # lowercase word, and removed white spaces
        
        # Split the word list by comma(,), then for each word => lowercase word, and removed white spaces
        keys = [ s.lower().strip() for s in data['Synonym(s)'].split(',')] 
        
        d = {key: value for key in keys } # Assign value to  list of keys
        
        synonym_dict.update(d)
    
```

```python
synonym_dict
```

```python
remove_tags_df = pd.read_excel('../mood_tags.xlsx', sheet_name='Remove ') # Read the remove tag dataframe
case_remove_tags = remove_tags_df['Remove'].tolist() # Convert the tags into python list
remove_tags = [tag.lower().strip() for tag in case_remove_tags] # lower case and remove white spaces
```

### This code prunes for the remove tags, null values and replace with synonyms

```python

Y = []
for row in raw_Y: # Iterate over each row
    y = []
    for element in row: # Iterate over each element in the row
        if not pd.isnull(element) and element.lower() not in remove_tags: #Filter based on condtion
            element = element.lower().strip()
            syn = synonym_dict.get(element, element) # get the synonym word from dict
            y.append(syn)
    y = list(set(y))
    Y.append(y)
            
```

```python
Y[:3] # You can observe there are multiple labels for each track
```

```python
import itertools
len(set(itertools.chain(*Y))) # To get the count of distinct words in the labels
```

```python
from sklearn.preprocessing import MultiLabelBinarizer # To encode multiple labels for the training
mlb = MultiLabelBinarizer(sparse_output=False)
# http://scikit-learn.org/stable/modules/multiclass.html#multilabel-classification-format
```

```python
mlb_y = mlb.fit(Y)
```

```python
multi_label_Y = mlb_y.transform(Y)
```

```python
multi_label_Y[0], len(multi_label_Y[0])
```

```python
from sklearn import linear_model
from sklearn.multiclass import OneVsRestClassifier
# http://scikit-learn.org/stable/auto_examples/plot_multilabel.html#multilabel-classification
```

```python
ridge_classifier_cv = linear_model.RidgeClassifierCV(cv=10) # Cross fold valiation helps you to understand weather your model is underfittnng or over fitting
```

```python
ridge_classifier_cv_model_OneVsRestClassifier = OneVsRestClassifier(ridge_classifier_cv).fit(X, multi_label_Y)
```

```python
#model = ridge_classifier_cv.fit(X, multi_label_Y)
def get_initial_result(model):
    result = model.predict(X)
    cnt = 0
    for (index, _), pred_res in zip(df.iterrows(), result):
        print(index, '   >>>>>>  ', mlb_y.inverse_transform(np.array([pred_res])))
        if cnt > 30:
            break
        cnt += 1
```
